 ### 1 Creating and Manipulating Arrays 

```python
import numpy as np
from matplotlib import pyplot as plt
```

### 1.1  Create two 2D arrays representing coordinates x, y on the cartesian plan

```python
x=np.linspace(-2,2,100)
y=np.linspace(-2,2,100)

xx,yy=np.meshgrid(x,y)
#print(xx)
#print (yy)
```

### 1.2 Visualize each 2D array using pcolormesh

```python
plt.pcolormesh(x,y,xx,cmap='viridis')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title ('Visualization of xx 2D array using pcolormesh')
```

```python
plt.title ('Visualization of yy 2D array using pcolormesh')
plt.pcolormesh(x,y,yy,cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

### From your cartesian coordinates, create polar coordinates r  an $\phi$


```python
r = np.sqrt(xx**2+yy**2)
phi=np.arctan2(yy,xx)
```

### 1.4. Visualize r and $\phi$ on the 2D x/y using pcolormeshmesh

```python
plt.pcolormesh(x,y,r,cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

```python
plt.pcolormesh(x,y,phi,cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

### 1.5 Caclulate the quanity f=cos2(4r)+sin2(4phi)

```python
f=(np.cos(4*r)**2)+(np.sin(4*phi)**2)
plt.pcolormesh(x,y,f,cmap='viridis')
plt.colorbar()
plt.xlabel('x')
plt.ylabel('y')
```

### 1.6 Plot the mean of f with respect to the x axis as a function of y

```python
plt.plot(x,f.mean(axis=0))
plt.xlabel('X')
plt.ylabel('Mean f')
```

```python
plt.plot(y,f.mean(axis=1))
plt.xlabel('y')
plt.ylabel('Mean f')
```

### 1.7 Plot the mean of f with respect to $\Phi$ as a function of r 


```python
r = np.sqrt(xx**2 + yy**2)
rmax = np.max(r)
r_range = np.linspace(0, rmax, num=100)
phi = np.arctan2(yy, xx)
phi_range = np.linspace(-np.max(phi), np.max(phi), num=100)
#print(phi_range)
r_grid, phi_grid = np.meshgrid(r_range, phi_range)
f = (np.cos(4*r_grid)**2) + (np.sin(4*phi_grid)**2)
plt.pcolormesh(r_grid, phi_grid, f, cmap='viridis')
plt.colorbar()
plt.xlabel('r')
plt.ylabel('$\phi$')
plt.show()
```

### 2.4 Compute the mean and standard deviation of each of T, S and P at each depth in level.

```python
T_mean=[]
T_std=[]
P_mean=[]
P_std=[]
S_mean=[]
S_std=[]
for i in range(len(level)):
    tt=np.mean(T[i,:])
    T_mean.append(tt)
    ts=np.std(T[i,:])
    T_std.append(ts)
    pt=np.mean(P[i,:])
    P_mean.append(pt)
    ps=np.std(P[i,:])
    P_std.append(ps)
    st=np.mean(S[i,:])
    S_mean.append(st)
    ss=np.std(S[i,:])
    S_std.append(ss)
plt.errorbar(T_mean,level,xerr=T_std)
plt.xlabel('Temperature')
plt.ylabel('Level')
plt.gca().invert_yaxis()
plt.title('Temperature vs Level with errorbars')
```

```python
plt.errorbar(S_mean,level,xerr=S_std)
plt.xlabel('Salinity')
plt.ylabel('Level')
plt.gca().invert_yaxis()
plt.title('Salinity vs Level with errorbars')
```

```python
plt.errorbar(P_mean,level,xerr=P_std)
plt.xlabel('Pressure')
plt.ylabel('Level')
plt.gca().invert_yaxis()
plt.title('Pressure vs Level')
```
