## 1. Reproduce the SST curve from the figure below
Use the sst.mnmean.nc file that we worked with in class, located at http://www.esrl.noaa.gov/psd/thredds/dodsC/Datasets/noaa.ersst.v5/sst.mnmean.nc.

You don’t have to match the stylistic details, or use different colors above and below zero, just the “3mth running mean” cur
Load the data as an Xarray dataset. Drop the time_bnds variable as we did in class and trim the data to 1950 onward for this assignment.
Now calculate the climatology and the SST anomaly

Now reproduce the plot. Keep the rolling 3-month average of the SST anomaly as a DataArray for the next question.ve.

```python
# Extract the SST data for the Niño 3.4 region from 1950 onwards
nino_sst = ds.sst.sel(lat=slice(5, -5), lon=slice(190, 240), time=slice('1950-01-01','2024-02-01'))
#Monthly anpmalies is given by fuction:
def anom(x):
    return (x - x.mean())
#monthly anomalies
nino_sst_anoms = nino_sst.groupby('time.month').apply(anom)
r_mean_anoms = nino_sst_anoms .rolling(time=3, center=True).mean() # r for running means
ONI = r_mean_anoms.mean(dim=('lat', 'lon'))
#Reproduce plot
plt.figure(figsize=(12, 6))
plt.plot(ONI.time, ONI, color='black', label='3-Month Running Mean', linewidth=1.25)
plt.fill_between(ONI.time, 0, ONI, 
                 where=ONI > 0, color='orange')
plt.fill_between(ONI.time, 0, ONI, 
                 where=ONI < 0, color='steelblue')
plt.axhline(y=0, color='black', linestyle='-')
plt.axhline(y=0.5, color= 'red', linestyle = 'dashed', alpha =0.4, label= 'El Nino Threshold')
plt.axhline(y=-0.5, color= 'blue', linestyle = 'dashed', alpha =0.4, label= 'La Nina Threshold')
plt.xlim(ONI.time.min(), ONI.time.max());
plt.title('SST Anomaly in Niño 3.4 Region (5N–5S, 120–170W)')
plt.xlabel('Year')
plt.ylabel('Anomaly in (°C)')
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 0.6),fancybox=True, ncol=1)
#plt.yticks(np.linspace(-2, 3, num=int((3 - (-2)) / 0.2) + 1))
```

## 2. Calculate boolean timeseries representing the positive / negative ENSO phases
Refer to the definitions of warm/cold phases above.

Plot them somehow.

```python
# Boolean masks for warm and cold phases
elnino_mask = ((ONI > 0.5).rolling(time=5, center=True).sum()==5).rolling(time=5, center=True, min_periods=1).max()
lanina_mask = ((ONI < -0.5).rolling(time=5, center=True).sum()==5).rolling(time=5, center=True, min_periods=1).max()
#Plot the boolean time series for warm and cold phases
plt.figure(figsize=(12, 6))
plt.fill_between(nino_sst.time, elnino_mask, color='blue', label='Positive')
plt.fill_between(nino_sst.time, lanina_mask, color='red', label='Negative')
plt.xlim(ONI.time.min(), ONI.time.max());
plt.ylim(elnino_mask.min(), elnino_mask.max());
plt.title('ENSO Phases')
plt.xlabel('Year')
plt.legend()
#plt.yticks([0, 1], ['False', 'True'])
plt.ylabel('Boolean Masks')
plt.legend(loc ='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=2)
plt.tight_layout()
```

## 3. Plot composites of SST anomaly for the positive and negative ENSO regimes
These should be pcolormesh maps. First positive ONI.

And negative ONI.

```python
sst = ds.sst.sel(time=slice('1950-01-01','2024-02-01'))
mean_sst = sst.mean(dim='time')
sst_anomalies = sst - mean_sst
p= sst_anomalies.where(elnino_mask) #P for positive ands N for negative
n= sst_anomalies.where(lanina_mask)
p_anomalies = p.mean(dim='time')
n_anomalies = n.mean(dim='time')
# Plot positive anoms
fig, ax = plt.subplots(figsize=(10, 10), nrows=2)
plot_p_anomalies = ax[0].pcolormesh(p_anomalies.lon, p_anomalies.lat, p_anomalies, vmin=-1.36, vmax=1.36, cmap='RdBu_r')
plt.colorbar(plot_p_anomalies, ax=ax[0], label='°C')
ax[0].set_title('SST Anomalies- Positive ENSO Phase')
ax[0].set_xlabel('Longitude [degrees_east]')  
ax[0].set_ylabel('Latitude [degrees_north]')
ax[0].grid(True, alpha=0.4)
#Plot negative anoms
plot_n_anomalies = ax[1].pcolormesh(n_anomalies.lon, n_anomalies.lat, n_anomalies, vmin=-1.36, vmax=1.36, cmap='RdBu_r')
plt.colorbar(plot_n_anomalies, ax=ax[1], label='°C')
ax[1].set_title('SST Anomalies- Negative ENSO Phase')
ax[1].set_xlabel('Longitude [degrees_east]')  
ax[1].set_ylabel('Latitude [degrees_north]')
ax[1].grid(True, alpha=0.4)
plt.tight_layout()
```

## 4. Calculate the composite of preciptiation for positive and negative ENSO phases.
First load the precip dataset. Code to fix the broken time coordinate is included.

```python
url = 'http://iridl.ldeo.columbia.edu/SOURCES/.NASA/.GPCP/.V2p1/.multi-satellite/.prcp/dods'
dsp = xr.open_dataset(url, decode_times=False)
true_time = (pd.date_range(start='1960-01-01', periods=len(dsp['T']), freq='MS'))
dsp['T'] = true_time
dsp = dsp.rename({'T': 'time'})
dsp.load()
```

## Now plot the difference between the time-mean of prcp during positive and negative ENSO phases.

```python
#get the precipitation data for warm and cold phases using boolean masks in q3
prcp_warm = dsp.prcp.where(elnino_mask)
prcp_cold = dsp.prcp.where(lanina_mask)
prcp_warm_mean = prcp_warm.mean(dim='time')
prcp_cold_mean = prcp_cold.mean(dim='time')
diff= prcp_warm_mean - prcp_cold_mean
# Plot
plt.subplots(figsize=(12, 6))
plot_difference = plt.pcolormesh(diff.X, diff.Y, diff, vmin=-1.75, vmax=1.75, cmap='RdBu_r')
plt.colorbar(plot_difference, label='mm/day')
plt.title('Difference between the time-mean of precipitation during positive and negative ENSO phases')
plt.xlabel('Longitude [degrees_east]')  
plt.ylabel('Latitude [degrees_north]')
plt.grid(True, alpha=0.4)
```
