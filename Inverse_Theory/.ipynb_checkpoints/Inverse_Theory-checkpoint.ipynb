```python
x = temp['time']
y= temp ['global_anom_mean']
plt.plot(x, y, marker='o',color= 'blue')
plt.show()
```

```python
temp = temp[["time", "global_anom_mean"]].set_index("time")
temp_resampled_monthly = temp.resample("ME")
temp = temp_resampled_monthly.mean().dropna(axis="index", how="any")
temp.plot()
plt.ylabel("Monthly Mean Temp")
plt.xlabel("Year")
_ = plt.title(
    "Monthly Mean Global Temperature Anomalies from NOAA 1850-2024")
```

```python
X = (temp.index.year + temp.index.month / 12).to_numpy().reshape(-1, 1)
y = temp["global_anom_mean"].to_numpy()
```

```python
from sklearn.gaussian_process.kernels import RBF
long_term_trend_kernel = 50.0**2 * RBF(length_scale=50.0)
```

```python
from sklearn.gaussian_process.kernels import ExpSineSquared
seasonal_kernel = (
    2.0**2
    * RBF(length_scale=100.0)
    * ExpSineSquared(length_scale=1.0, periodicity=1.0, periodicity_bounds="fixed"))
```

```python
from sklearn.gaussian_process.kernels import RationalQuadratic
irregularities_kernel = 0.5**2 * RationalQuadratic(length_scale=1.0, alpha=1.0)
```

```python
from sklearn.gaussian_process.kernels import WhiteKernel
noise_kernel = 0.1**2 * RBF(length_scale=0.1) + WhiteKernel(
    noise_level=0.1**2, noise_level_bounds=(1e-5, 1e5))
```

```python
temp_kernel = (long_term_trend_kernel + seasonal_kernel + irregularities_kernel + noise_kernel)
temp_kernel
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
y_mean = y.mean()
gaussian_process = GaussianProcessRegressor(kernel=temp_kernel, normalize_y=False)
gaussian_process.fit(X, y - y_mean)
```

```python
import datetime
today = datetime.datetime.now()
#current_month = today.year + today.month / 12
X_test = np.linspace(start=1850, stop=2030, num=1_000).reshape(-1, 1)
mean_y_pred, std_y_pred = gaussian_process.predict(X_test, return_std=True)
mean_y_pred += y_mean
```

```python
plt.figure(figsize=(15, 9))
plt.plot(X, y, color="blue", linestyle="solid", marker= 'o', label="Data Measurements",alpha =0.4)
plt.plot(X_test, mean_y_pred, color="red", label="Gaussian process predictions")
plt.axhline(y=0, color='black', linestyle='-', linewidth = 1.5)
plt.legend()
plt.xlabel("Year")
plt.ylabel("Temp Anomalies")
_ = plt.title("Monthly Mean Global Temp Anomalies from NOAA 1850-2024")
plt.savefig('MGTA_GPR_OBS.tiff', format='tiff')
```

```python
# Calculate upper and lower bounds of the uncertainty envelope
upper_bound = mean_y_pred + 2 * std_y_pred
lower_bound = mean_y_pred - 2 * std_y_pred

# Plot the data measurements, predictions, and uncertainty envelope
plt.figure(figsize=(15, 9))
plt.plot(X, y, color="blue", linestyle="solid", marker='o', label="Data Measurements", alpha=0.4)
plt.plot(X_test, mean_y_pred, color="red", label="Gaussian process predictions")
plt.fill_between(X_test.flatten(), upper_bound, lower_bound, color="gray", alpha=0.7, label="Uncertainty Envelope")
plt.axhline(y=0, color='black', linestyle='-', linewidth=1.5)
plt.legend(loc = 'upper left')
plt.xlabel("Year")
plt.ylabel("Temp Anomalies")
_ = plt.title("Monthly Mean Global Temp Anomalies from NOAA 1850-2024")

# Save plot as TIFF
plt.savefig('MGTA_GPR_OBS_with_uncertainty.tiff', format='tiff')
```

44.8**2 * RBF(length_scale=51.6) + 2.64**2 * RBF(length_scale=91.5) * ExpSineSquared(length_scale=1.48, periodicity=1) + 0.536**2 * RationalQuadratic(alpha=2.89, length_scale=0.968) + 0.188**2 * RBF(length_scale=0.122) + WhiteKernel(noise_level=0.0367)


Thus, most of the target signal, with the mean subtracted, is explained by a long-term rising trend for ~45 ppm and a length-scale of ~52 years. The periodic component has an amplitude of ~2.6ppm, a decay time of ~90 years and a length-scale of ~1.5. The long decay time indicates that we have a component very close to a seasonal periodicity. The correlated noise has an amplitude of ~0.2 ppm with a length scale of ~0.12 years and a white-noise contribution of ~0.04 ppm. Thus, the overall noise level is very small, indicating that the data can be very well explained by the model.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sine wave data
x = np.linspace(0, 2 * np.pi, 100)  # 100 points between 0 and 2*pi
y = np.sin(x)

# Select five points on the sine wave
points_x = [x[10], x[30], x[50], x[70], x[90]]
points_y = [y[10], y[30], y[50], y[70], y[90]]
point_labels = ['X1', 'X2', 'X3', 'X4', 'X5']

# Calculate Euclidean distances between points X1-X2 and X4-X5
euclidean_dist_12 = np.sqrt((points_x[1] - points_x[0])**2 + (points_y[1] - points_y[0])**2)
euclidean_dist_45 = np.sqrt((points_x[4] - points_x[2])**2 + (points_y[4] - points_y[2])**2)

# Calculate slopes for tangents
slope_12 = (points_y[1] - points_y[0]) / (points_x[1] - points_x[0])
slope_45 = (points_y[4] - points_y[2]) / (points_x[4] - points_x[2])

# Plot the sine wave
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sine Wave')

# Plot the selected points
plt.scatter(points_x, points_y, color='red')
for i, label in enumerate(point_labels):
    plt.text(points_x[i], points_y[i], label, fontsize=12, ha='right', va='bottom')

# Plot tangents
x_tangent_12 = np.linspace(points_x[0], points_x[1], 100)
y_tangent_12 = slope_12 * (x_tangent_12 - points_x[0]) + points_y[0]
plt.plot(x_tangent_12, y_tangent_12, linestyle='--', color='blue', label=f'Tangent X1-X2\nDistance: {euclidean_dist_12:.2f}')

x_tangent_45 = np.linspace(points_x[2], points_x[4], 100)
y_tangent_45 = slope_45 * (x_tangent_45 - points_x[2]) + points_y[2]
plt.plot(x_tangent_45, y_tangent_45, linestyle='--', color='green', label=f'Tangent X4-X5\nDistance: {euclidean_dist_45:.2f}')

# Label axes and add legend
plt.xlabel('X')
plt.ylabel('Y')



# Show plot

plt.show()
```
