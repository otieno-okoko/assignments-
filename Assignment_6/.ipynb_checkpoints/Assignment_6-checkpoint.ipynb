## 15) Calculate the climatology of datapoint counts as a function of day of year
Plot the mean and standard deviation on a single figur

Using WIND as our Climatology Proxy having removed NaNs we get the following result.

e

```python
NA_time['Date']=pd.to_datetime(NA_time.index).date
NA_time['DOY'] = pd.to_datetime(NA_time['Date']).dt.dayofyear
MStd=NA_time.groupby('DOY').agg({'WIND':['mean','std']}) #MStd is M=mean Std=Standard deviation
fig,(ax1)=plt.subplots(1,1,figsize=(12,10),sharex=True)
ax1.plot(MStd.index,MStd['WIND']['mean'],label='Mean')
ax1.plot(MStd.index,MStd['WIND']['std'],label='Standard Deviation')
ax1.legend()
ax1.set_ylabel('Mean/Standard Deviation')
ax1.set_xlabel('Day of the Year)')
```

## 16) Use transform to calculate the anomaly of daily counts from the climatology
Resample the anomaly timeseries at annual resolution and plot a line with dots as markers.

Which years stand out as having anomalous hurricane activity?

```python
def standardize(x):
    return (x - x.mean())/x.std()
    
basin=NA_time[['DOY','SEASON']]
trans_Std=basin.groupby('DOY').transform(standardize)
resample_Ye=trans_Std.resample('YE').mean()
resample_Ye.plot(y=['SEASON'],marker='o')
plt.xlabel('Year');
plt.ylabel('Anomaly');
print("These are the anomalous years when above 0.3 or below -0.3:")
print(resample_Ye.index[((resample_Ye['SEASON'] > 0.3) | (resample_Ye['SEASON'] < -0.3))].year)

```
